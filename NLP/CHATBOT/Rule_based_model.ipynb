```python
import tkinter as tk
from tkinter import messagebox

# Configuraci칩n inicial
puntaje = 0
preguntas = [
"Is the lesion asymmetrical in shape (not round or oval)?",
"Are the edges of the lesion irregular or fuzzy?",
"Is the lesion multi-colored (brown, black, red, white)?",
"Is the diameter of the lesion greater than 6 mm (like a pencil eraser)?",
"Has the lesion changed in size, shape, or color recently?"
]
indice_pregunta = 0

# Funci칩n para manejar respuestas
def responder(respuesta):
    global puntaje, indice_pregunta
    
    if respuesta:
        puntaje += 1
    
    indice_pregunta += 1
    
    if indice_pregunta < len(preguntas):
        label_pregunta.config(text=preguntas[indice_pregunta])
        label_progreso.config(text=f"Question {indice_pregunta + 1}/{len(preguntas)}")
    else:
        mostrar_resultado()

# Funci칩n para mostrar el resultado
def mostrar_resultado():
    # Ocultar elementos anteriores
    label_pregunta.pack_forget()
    boton_si.pack_forget()
    boton_no.pack_forget()
    label_progreso.pack_forget()
    
    # Mostrar resultado
    resultado_texto = f"Score ABCDE: {puntaje}/5\n"
    if puntaje >= 3:
        resultado_texto += "游댮 HIGH RISK of melanoma. Consult a dermatologist."
        color = "#c0392b"
    elif puntaje >= 1:
        resultado_texto += "游리 MODERATE RISK. Monitor the injury and schedule a follow-up."
        color = "#f39c12"
    else:
        resultado_texto += "游릭 LOW RISK. Continue monitoring the injury for changes."
        color = "#27ae60"
    
    label_resultado = tk.Label(
        root,
        text=resultado_texto,
        font=("Arial", 12, "bold"),
        fg=color,
        bg="#f0f8ff",
        wraplength=400,
        justify="center"
    )
    label_resultado.pack(pady=50)
    
    boton_cerrar = tk.Button(
        root,
        text="Cerrar",
        command=root.destroy,
        bg="#3498db",
        fg="white",
        font=("Arial", 12)
    )
    boton_cerrar.pack()

# Crear ventana principal
root = tk.Tk()
root.title("Chatbot OF Melanoma")
root.geometry("500x400")
root.configure(bg="#f0f8ff")

# T칤tulo
titulo = tk.Label(
    root,
    text="Assess ABCDE of Melanoma",
    font=("Arial", 16, "bold"),
    bg="#f0f8ff",
    fg="#2e86c1"
)
titulo.pack(pady=20)

# Pregunta actual
label_pregunta = tk.Label(
    root,
    text=preguntas[indice_pregunta],
    font=("Arial", 12),
    bg="#f0f8ff",
    wraplength=400,
    justify="center"
)
label_pregunta.pack(pady=20)

# Botones de respuesta
frame_botones = tk.Frame(root, bg="#f0f8ff")
frame_botones.pack(pady=10)

boton_si = tk.Button(
    frame_botones,
    text="yes",
    command=lambda: responder(True),
    bg="#2ecc71",
    fg="white",
    font=("Arial", 12),
    padx=20
)
boton_si.pack(side="left", padx=10)

boton_no = tk.Button(
    frame_botones,
    text="no",
    command=lambda: responder(False),
    bg="#e74c3c",
    fg="white",
    font=("Arial", 12),
    padx=20
)
boton_no.pack(side="right", padx=10)

# Barra de progreso
label_progreso = tk.Label(
    root,
    text=f"Question {indice_pregunta + 1}/{len(preguntas)}",
    bg="#f0f8ff",
    font=("Arial", 10)
)
label_progreso.pack(pady=20)

root.mainloop()
```
