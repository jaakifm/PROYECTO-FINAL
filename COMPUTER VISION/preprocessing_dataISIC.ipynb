```python
# Directory where images are stored
image_dir = r"C:\PROYECTO_FINAL\images\ISIC-images(bcn20000)"

# convert into a list all the images
image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg')]

# Cargar las primeras 5 imgenes
train_images = []
for i in range(5):
    img_path = os.path.join(image_dir, image_files[i])
    img = Image.open(img_path)
    train_images.append(np.array(img))

# Mostrar las primeras 5 imgenes
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(train_images[i])
    plt.axis('off')
plt.show()
```
